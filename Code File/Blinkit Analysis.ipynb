## **Business Requirements**

## **KPI's Requirements**

```python
# Total Sales
Total_Sales = df['Sales'].sum()

#Average Sales
Average_Sales = df['Sales'].mean()

#Number of Items Sold
no_of_items_sold = df['Sales'].count()

#Average Rating 
Average_Rating = df['Rating'].mean()

#Display

print(f"Total_Sales: ${Total_Sales: ,.0f}")

print(f"Average_Sales: ${Average_Sales: ,.1f}")

print(f"Number of Items Sold: ${no_of_items_sold: ,.0f}")

print(f"Average Rating: ${Average_Rating: ,.1f}")
```

## **Charts Requirements**

### **Total Sales by Fat Content**

```python
sns.set_style('ticks')
Sales_by_Fat_Content = df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(Sales_by_Fat_Content , labels = Sales_by_Fat_Content.index , autopct = "%0.1f%%" , startangle = 90 , colors = ['purple' , 'gold'])
plt.title('Sales by Fat Content' , fontsize = 15)
```

### **Total Sales by Item Type**

```python
sns.set_style('darkgrid')
Sales_by_Item_Type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending = False)
plt.figure(figsize = (10 , 6))
plt.bar(Sales_by_Item_Type.index , Sales_by_Item_Type.values , color = 'purple')
plt.xticks(rotation = 90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Sales by Item Type' , fontsize = 15)
```

### **Fat Content by Outlet for Total Sales**

```python
sns.set_style('darkgrid')
grouped = df.groupby(['Outlet Location Type' , 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular' , 'Low Fat']]
grouped.plot(kind = 'bar' , color = ['gold','purple'])
plt.title('Outlet Type by Item Fat Content' , fontsize = 15)
plt.ylabel('Total Sales')
```

### **Total Sales by Outlet Establishment**

```python
Sales_Outlet_Establishment = df.groupby('Outlet Establishment Year')['Sales'].sum()
sns.lineplot(x = Sales_Outlet_Establishment.index , y = Sales_Outlet_Establishment.values , marker = 'o' , color = 'purple')
plt.title('Sales by Outlet Establishment Year' , fontsize = 15)
```

### **Sales by Outlet Size**

```python
Sales_Outlet_Size = df.groupby('Outlet Size')['Sales'].sum().sort_values(ascending = False)
plt.pie(Sales_Outlet_Size , labels = Sales_Outlet_Size.index , autopct = '%0.1f%%' , colors = ['purple' , 'gold' , 'darkblue'])
plt.title('Sales by Outlet Size' , fontsize = 15)
```

### **Sales by Outlet Location**

```python
Sales_Outlet_Location = df.groupby('Outlet Location Type')['Sales'].sum()
sns.barplot(x = Sales_Outlet_Location.values , y = Sales_Outlet_Location.index , color = 'purple')
plt.title('Sales by Outlet Location' , fontsize = 15)
```
